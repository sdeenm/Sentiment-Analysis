# Step 3: Exploratory Analysis 

```python
!pip install wordcloud
```

```python
# Read a text file and calculate frequency of words in it

words = papers.split()

#removing stop words
with open("stopwords_en.txt", "r") as f:
    stop_words = f.read().split()

data = dict()

for word in words:
    word = word.lower()
    if word in stop_words:
        continue
    data[word] = data.get(word, 0) + 1
```

```python
data
```

```python
# image configurations
background_color = "#101010"
height = 720
width = 1080
```

```python
from wordcloud import WordCloud 

word_cloud = WordCloud(
    background_color=background_color,
    width=width,
    height=height
)

word_cloud.generate_from_frequencies(data)
word_cloud.to_file('word_cloud_image.png')
```

# Step 4: Prepare text for LDA analysis 

We convert texts into tokens and convert the tokenized object into a corpus and dictionary.

```python
import nltk
nltk.download('punkt')
tokens = nltk.word_tokenize(papers)
print (tokens)
```

```python
words
```

```python
import gensim.corpora as corpora
# Create Dictionary
dictionary = corpora.Dictionary([tokens])
print(dictionary)
```

```python
print(dictionary.token2id)
```

```python
# Create Corpus
#texts = data_words
# Term Document Frequency
corpus = [dictionary.doc2bow(tokens) for doc in papers]
# View
#print(corpus)
```

```python
import gensim
from pprint import pprint
# number of topics
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=dictionary,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
import os
import pyLDAvis.gensim
import pickle 
import pyLDAvis
# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_data_filepath = os.path.join('./results/ldavis_prepared_'+str(num_topics))
# # this is a bit time consuming - make the if statement True
# # if you want to execute visualization prep yourself


if 1 == 1:
    LDAvis_prepared = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)
        
        
# load the pre-prepared pyLDAvis data from disk
with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared, './results/ldavis_prepared_'+ str(num_topics) +'.html')
#pyLDAvis.save_html(LDAvis_prepared, lda.html)
LDAvis_prepared
```
